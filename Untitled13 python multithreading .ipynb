```python

from threading import *
from time import sleep



class naga(Thread):
    def run(self):
        for i in range(50):
            print("naga")
            sleep(1)
            
class siri(Thread):
    def run(self):
        for i in range(55):
            print("siri")
            sleep(1.5)
            
t1=naga()
t2=siri()


# t1.run()
# t2.run()


t1.start()     #  start() method is used for thread concept 
t2.start()

```

```python
from threading import *



class naga(Thread):
    def run(self):
        for i in range(50):
            print("naga")
            
class siri(Thread):
    def run(self):
        for i in range(55):
            print("siri")
            
t1=naga()
t2=siri()


# t1.run()
# t2.run()


t1.start()     #  start() method is used for thread concept 
t2.start()

t1.join()
t2.join()


print(" hello : now it is print at last ")
```
